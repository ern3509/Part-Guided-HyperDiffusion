```python
import os
import torch


input_dir = "logs/insert_name_here_with_part_head"       # where the overfit MLPs with part heads are
output_dir = "logs/insert_name_here_without"         # where to save new MLPs without part heads
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if not file.endswith(".pth"):
        continue

    checkpoint_path = os.path.join(input_dir, file)
    state_dict = torch.load(checkpoint_path, map_location='cpu')

    # Remove part heads (class_heads)
    filtered_state_dict = {
        k: v for k, v in state_dict.items() if not k.startswith("class_heads.")
    }

    # Save to new file
    save_path = os.path.join(output_dir, file)
    torch.save(filtered_state_dict, save_path)
    print(f"✅ Saved: {save_path}")
```
